```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import military spending dataset, getting rid of header and footer
milsp = pd.read_excel('military spending.xlsx', skiprows=5, skipfooter=8, index_col=0)
milsp.head()
```

```python
#Get my 10 countries
mycountries = milsp.loc[['USA', 'China, P.R.', 'Russian Federation', 'Germany', 'UK', 'France', 'Italy', 'Saudi Arabia', 'Korea, South', 'Australia']]
```

```python
mycountries
```

```python
#get appropriate years of data for 10 countries stored in new dataframe
mycountriesms = mycountries.iloc[:,65:71] #select correct columns (years)
mycountriesms
```

```python
#Graph and compare military spending through the years amongst the countries
milsplot = mycountriesms.T.plot(figsize=(20,10), fontsize=13)
plt.legend(loc='best', fontsize=13) #best is default fitted
milsplot.set_title('Military Spending Among Countries')
milsplot.set_xlabel('Year', fontsize=13)
milsplot.set_ylabel('Military Spending (mil. USD)', fontsize=13)
plt.show()
#Notice how much more USA spends than everyone
```

```python
#Bar Graph
milbar = mycountriesms.T.plot(kind='bar', figsize=(10,6), title = 'Military Spending')
milbar.set_xlabel('Year', fontsize=13)
milbar.set_ylabel('Military Spending (mil. USD)', fontsize=13)
plt.show()
#We can see that the USA outspends all other countries by significant amount on military
```

```python
#Read in military spending as % of GDP
perGDP = pd.read_excel('share of GDP.xlsx', skiprows=5, skipfooter=8, index_col=0)
perGDP.tail()
```

```python
#Get the 10 countries
GDPshare = perGDP.loc[['USA', 'China, P.R.', 'Russian Federation', 'Germany', 'UK', 'France', 'Italy', 'Saudi Arabia', 'Korea, South', 'Australia']]
```

```python
GDPshare
```

```python
#get appropriate years (2012-2017)
GDPsharef = GDPshare.iloc[:,64:71] #select correct columns (years)
GDPsharef
```

```python
#Plot for GDP %
gdpplt = GDPsharef.T.plot(figsize=(20,10), fontsize=13)
plt.legend(loc='best', fontsize=13) #best is default fitted
gdpplt.set_title('Military Spending as Percent of GDP')
gdpplt.set_xlabel('Year', fontsize=13)
gdpplt.set_ylabel('% of GDP Spent on Military (in dec)', fontsize=13)
plt.show()
#Saudi Arabia has highest % of GDP spent on military by considerable margins
```

```python
#Bar Graph
GDPbar = GDPsharef.T.plot(kind='bar', figsize=(10,6), title = 'Military Spending as Percent of GDP')
GDPbar.set_xlabel('Year', fontsize=13)
GDPbar.set_ylabel('% of GDP Spent on Military(in dec)', fontsize=13)
plt.show()
```

```python
#Combined military spending by year
totmilspend = mycountriesms.sum()
totmilspend
```

```python
#Same thing but total for each country within the range of years
totmilspendc = mycountriesms.sum(axis=1)
totmilspendc
```

```python
#Plot total, combined military spending throughout the years
tmsplt = totmilspend.T.plot(figsize=(20,10), fontsize=13)
tmsplt.set_title('Historical Total Military Spending (US, Russia, China, UK, France, China, Italy, Saudi Arabia, Australia, S Korea)')
tmsplt.set_xlabel('Year', fontsize=13)
tmsplt.set_ylabel('Military Spending (mil. USD)', fontsize=13)
plt.show()
```

```python
#Bar Graph
totmsbar = totmilspendc.plot(kind='barh', figsize=(10,6), title = 'Total Military Spending (2012-2017)')
totmsbar.set_xlabel('Military Spending (mil. USD)', fontsize=13)
totmsbar.set_ylabel('Country', fontsize=13)
plt.show()
#USA has by far the most spent on military over the years
```

```python
#Next compare per person military spending to per person GDP
#We have excel file on military spending per capita
#Read in per capita spending 
percapitams = pd.read_excel('military spending per capita.xlsx', skiprows=6, skipfooter=7, index_col=0)
percapitams.tail()
```

```python
#Locate the 10 countries
mspercap = percapitams.loc[['USA', 'China, P.R.', 'Russian Federation', 'Germany', 'UK', 'France', 'Italy', 'Saudi Arabia', 'Korea, South', 'Australia']]
mspercap
```

```python
#Get 2012-2017 data
mspercapita = mspercap.iloc[:,25:31] #select correct columns (years)
mspercapita
```

```python
#Need to calculate population to find GDP per capita
#per person * pop = total spent--use algebra
population = (mycountriesms) / (mspercapita)
population
#population is in millions
```

```python
#Also need total GDP for each country
#Total GDP multiplied by the share that was spent on military equals total military spending. 
GDPtot = (mycountriesms) / (GDPsharef)
GDPtot
```

```python
#Finally, calculate GDP per capita
GDPpercapita = (GDPtot) / (population)
GDPpercapita
```

```python
#Plot Military Spending Per Capita
mspercapplt = mspercapita.T.plot(figsize=(20,10), fontsize=13)
mspercapplt.set_title('Military Spending Per Capita (US, Russia, China, UK, France, China, Italy, Saudi Arabia, Australia, S Korea)')
mspercapplt.set_xlabel('Year', fontsize=13)
mspercapplt.set_ylabel('Military Spending Per Capita (USD)', fontsize=13)
plt.show()
```

```python
#Plot GDP Per Capita
GDPpercapitaplt = GDPpercapita.T.plot(figsize=(20,10), fontsize=13)
GDPpercapitaplt.set_title('GDP Per Capita (US, Russia, China, UK, France, China, Italy, Saudi Arabia, Australia, S Korea)')
GDPpercapitaplt.set_xlabel('Year', fontsize=13)
GDPpercapitaplt.set_ylabel('GDP Per Capita (USD)', fontsize=13)
plt.show()
```

```python
#Fixed Value Growth for 2012-2017
mycountriesms['Fixed Value'] = mycountriesms[2017] - mycountriesms[2012]
mycountriesms
```

```python
#Military Spending Growth as a percentage ((new-old)/(old))
mycountriesms ['Growth Percentage'] = (mycountriesms['Fixed Value']/mycountries[2012])*100
mycountriesms
```

```python
#Fixed Value Growth Graph
FixedBar = mycountriesms.iloc[:,-2].plot(kind='bar', figsize=(10,6), title = 'Fixed Value Military Spending Growth (2012-2017)')
FixedBar.set_xlabel('Country', fontsize=13)
FixedBar.set_ylabel('Fixed Value Growth (USD mil)', fontsize=13)
plt.show()
```

```python
#Percentage Growth Graph
PerGrowthBar = mycountriesms.iloc[:,-1].plot(kind='bar', figsize=(10,6), title = 'Percentage Military Spending Growth (2012-2017)')
PerGrowthBar.set_xlabel('Country', fontsize=13)
PerGrowthBar.set_ylabel('Percentage Growth', fontsize=13)
plt.show()
```
